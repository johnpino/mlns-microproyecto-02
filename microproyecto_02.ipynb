# Alcanzando los Objetivos de Desarrollo Sostenible
## Un aporte desde el machine learning

La Organización de las Naciones Unidas (ONU) adoptó el 25 de septiembre de 2015 la Agenda 2030 para el Desarrollo Sostenible, cuyo propósito es reducir la pobreza, garantizar acceso a salud y educación, promover la igualdad de género y oportunidades, y disminuir el impacto ambiental, entre otros objetivos.

La Agenda 2030 define **17 Objetivos de Desarrollo Sostenible (ODS)** y **169 metas**. En este contexto, entidades como el Fondo de Población de las Naciones Unidas (UNFPA), en articulación con instituciones públicas y herramientas de participación ciudadana, buscan identificar problemáticas y evaluar soluciones actuales relacionando información con los diferentes ODS.

Un reto clave es la interpretación y análisis de información textual proveniente de distintas fuentes de planeación participativa territorial. Esta tarea requiere gran cantidad de recursos y expertos que relacionen textos con ODS. Contar con un sistema automatizado facilitaría la toma de decisiones basada en la opinión de la población y permitiría orientar mejor las políticas públicas hacia el cumplimiento de la Agenda 2030.

El objetivo del proyecto es desarrollar una solución basada en procesamiento de lenguaje natural y machine learning que permita clasificar automáticamente textos según los 17 ODS, ofreciendo resultados en una forma comprensible para el usuario final.


## A. Objetivo

- Desarrollar una solución, basada en técnicas de procesamiento de lenguaje natural y machine learning, que facilite la interpretación y análisis de información textual para identificar relaciones semánticas con los Objetivos de Desarrollo Sostenible.


## B. Conjunto de datos

El conjunto de datos forma parte del proyecto **OSDG Community Dataset (OSDG-CD)** (versión 2023), con un total de **40.067 textos**, de los cuales **3.000** provienen de fuentes relacionadas con Naciones Unidas.

También incluye documentos públicos, resúmenes de artículos y reportes. La plataforma reúne investigadores, expertos y defensores de los ODS de todo el mundo para crear una fuente amplia y precisa de información textual sobre ODS. Los voluntarios validan la relevancia de cada texto para los ODS mediante ejercicios de etiquetado.

Los textos usados en este proyecto fueron traducidos al español con herramientas como Deepl y se realizó aumentación de textos a través del API de ChatGPT.


## C. Actividades para realizar

1. Preparar los textos utilizando el esquema de **bolsa de palabras (BOW)** con ponderación **TF-IDF**. Para este paso, construir un pipeline que integre las transformaciones que se consideren adecuadas.
2. Desarrollar un modelo de clasificación que permita relacionar un texto con un ODS. Para manejar la complejidad del espacio de entrada, aplicar un algoritmo de reducción de dimensionalidad.
3. Evaluar el modelo con textos que no hayan sido utilizados para el aprendizaje.


## D. Consideraciones

- El algoritmo de clasificación y la técnica de reducción de dimensionalidad quedan a consideración de cada grupo, pero es importante justificar la elección.
- Es posible utilizar otra técnica para la representación vectorial de los textos, como **Word2Vec**.


## E. Entregable

- Entregar notebook en formato **`.ipynb`** y **`.html`** del método desarrollado.
- El notebook debe estar documentado con justificaciones de las decisiones tomadas en cada paso.
- Deben ser visibles las ejecuciones de cada celda.
- Para evidenciar el desempeño, el notebook debe mostrar clasificaciones para **al menos cuatro textos** del conjunto de prueba.

La entrega se realiza al final de la **semana 7**, adjuntando ambos archivos.


## F. Criterios de evaluación

| Actividad | Porcentaje |
|---|---:|
| Preparación de los datos, incluyendo reducción de dimensionalidad, justificando decisiones | 35% |
| Construcción del pipeline de preparación de datos | 15% |
| Construcción del modelo de clasificación con búsqueda de hiperparámetros, validación con métricas adecuadas, justificación de algoritmo y métricas, y aplicación de reducción de dimensionalidad | 35% |
| Evidencia del desempeño mostrando clasificaciones sobre textos no usados en aprendizaje | 15% |


## G. Bonificación adicional (opcional)

Con el objetivo de fortalecer competencias en despliegue y aplicación práctica de modelos de machine learning, se otorgará una bonificación adicional de **15 puntos** a los grupos que implementen voluntariamente el modelo en una aplicación interactiva usando **Streamlit**.

La aplicación deberá:

- Permitir al usuario ingresar un texto libre.
- Procesar el texto con el mismo pipeline construido en el proyecto.
- Generar como salida la predicción del ODS correspondiente.
- Ser funcional y ejecutarse correctamente.

Esta bonificación es opcional y no reemplaza los criterios de la rúbrica; corresponde a un reconocimiento adicional por llevar el modelo a un escenario de uso práctico.
